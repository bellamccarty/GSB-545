Welcome to your **"Choose Your Own Deep Learning Adventure"**! This final project gives you the flexibility to explore deep learning topics of your choosing, while still grounding your work in modern Keras/TensorFlow practices.

## Project Goals

- **Explore deep learning topics not fully covered in the course.**
- **Create two new, well-documented Keras examples** that follow modern best practices.
- **Pitch deep learning as a business analytics tool**, targeting real-world applications.

---

## Project Requirements

You will submit **two separate Jupyter notebooks**, each containing one of your custom Keras examples. Both should follow the structure and style found on [keras.io/examples](https://keras.io/examples/), with the following considerations:

### Each Example Should: https://keras.io/examples/structured_data/movielens_recommendations_transformers/

- Use a **new dataset** (different from the original Keras example, and different between your two examples).
- Be **shorter than 300 lines of code**, *excluding data wrangling and preprocessing*.
- Be **clearly and thoroughly commented**.
- Include **rich explanatory narrative text** that makes the notebook accessible to students who have completed GSB 544.
- Reflect **modern Keras/TensorFlow best practices**, mirroring those seen in official examples.

You are **not required** to write your model from scratch. Instead, feel free to:

- Select existing Keras examples and reimplement them using a **new dataset** and **improved narrative**.
- Modify architectures, loss functions, or evaluation metrics where relevant.
